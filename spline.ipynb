```python
import numpy as np
import pylab as pl
from scipy import interpolate
import pylab as pl

air = np.loadtxt('/content/drive/MyDrive/Plasma_data/spectrum/spectrumN2Air.txt')
N2 = np.loadtxt('/content/drive/MyDrive/Plasma_data/spectrum/spectrumN2Air2.txt')

f_linear = interpolate.interp1d(air[:,0], air[:,1])

pl.plot(N2[:,0], f_linear(N2[:,0])-N2[:,1], label=u"Linear interpolation")

f = open('/content/drive/MyDrive/Plasma_data/spectrum/AMMM.txt', "w")
for i in range(0, len(N2)):
  f.write(str(N2[i,0]) + '\t' + str(f_linear(N2[i,0])-N2[i,1]) + '\n')
  #print(str(N2[i,0]) + '\t' + str(f_linear(N2[i,0])-N2[i,1]) + '\n')
f.close
```
